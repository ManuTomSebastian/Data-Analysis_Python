**3. Display the percentage distribution of the data in each year using the pie chart.**

```python
my_labels=data1['year']
my_labels.drop_duplicates(inplace=True)
print(my_labels)
```

```python
df=data1['year'].value_counts()
df
```

```python
plt.figure(figsize=(5, 5), facecolor='white') 
plt.pie(df, labels=my_labels, autopct='%.2f%%', labeldistance=1.02, startangle=10, colors=sns.color_palette('Set3'), shadow=True)
plt.title('Year Distribution')
plt.axis('equal')
plt.show()
```

Here year value count is taken to obtain the year distribution.

In the above Pie chart we can see that the highest distribution is "7.03%" and the years are "1998, 1999 & 2000". The honey production supply and demand was high in these years.

And the least distribution is "6.39%" and the years are "2009, 2010, 2011 & 2012". The honey production supply and demand was less in these years.

**5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and
write your findings.**

```python
sns.scatterplot(x=data1['numcol'], y=data1['prodvalue'], data=data1)
plt.title('numcol vs prodvalue')
plt.xlabel('numcol')
plt.ylabel('prodvalue')
plt.show()
```

Here we can say that the relationship between "numcol" and "prodvalue" is positive and linear. That is the production values increase with the high increase in number of honey producing colonies. 

There are some outliers in the dataset.
