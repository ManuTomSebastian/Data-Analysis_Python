**8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.**

```python
data2['video_count']=1
df1 = pd.DataFrame(data2, columns=['published_month','video_count'])
Video_Count = pd.pivot_table(df1, index=['published_month'],aggfunc='sum')
Video_Count.sort_values('video_count',ascending=False)
```

**13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.**

```python
df7 = pd.DataFrame(data2, columns=['view_count','dislikes'])
df7.cov()
```

Here we can see a very high covariance of "5.118389e+11" between view_count & dislikes. It's a positive value and also have a strong relationship between these two variables. 


```python
df7.corr()
```

Here we can see a high positive correlation of "0.684" which is close to 1, indicating a significant relationship between view_count & dislikes. When one variable increase, the other increases too.

```python
sns.pairplot(df7, kind='reg')
plt.show()
```

Here in the above scatterplot of view_count & dislikes we can see that the dislikes increases slowly with respect to high increase in view_count. Therefore we can say the number of views have a small effect on how many people disliked the video.
